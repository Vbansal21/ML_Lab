Objective: Implement a decision tree classifier using scikit-learn. Tune the hyperparameters of the decision tree. Visualize the constructed decision tree.

```python
import os
import shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import kagglehub

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

###############################################################################
# 1) Downloading and loading dataset
###############################################################################
def download_data():
    """
    Downloads the Spotify dataset from Kaggle, looks for 'data.csv', and loads it into a pandas DataFrame.
    """
    print("Downloading dataset...")
    path = kagglehub.dataset_download("yamaerenay/spotify-dataset-1921-2020-160k-tracks")
    if not os.path.exists('data.csv'):
        for root, dirs, files in os.walk(path):
            if 'data.csv' in files:
                shutil.copy2(os.path.join(root, 'data.csv'), 'data.csv')
                break
    if os.path.exists('data.csv'):
        print("Dataset downloaded successfully!")
        return pd.read_csv('data.csv')
    else:
        raise FileNotFoundError("data.csv not found after download.")

df = download_data()

###############################################################################
# 2) Feature selection & minimal cleaning
###############################################################################
# We select the same features as Lab 3 but will cast the 'popularity' (continuous)
# into a binary classification target. For example, let's say anything with popularity
# >= 60 is "popular" (label = 1) and below 60 is "not popular" (label = 0).
features = [
    'acousticness', 'danceability', 'duration_ms', 'energy',
    'instrumentalness', 'liveness', 'loudness', 'mode',
    'speechiness', 'tempo', 'valence'
]
target = 'popularity'

# Drop rows with NaNs in features or target
df = df.dropna(subset=features + [target])

X = df[features].values
y_binary = (df[target] >= 60).astype(int)  # Binary classification target

###############################################################################
# 3) Train-test split & scaling
###############################################################################
X_train, X_test, y_train, y_test = train_test_split(
    X, y_binary, test_size=0.2, random_state=42
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

###############################################################################
# 4) Decision Tree Classifier Implementation
###############################################################################
# We'll tune hyperparameters like 'max_depth', 'min_samples_split', and 'criterion'.

# Step 4.1: Define the parameter grid for GridSearchCV
param_grid = {
    'criterion': ['gini', 'entropy'],     # Which impurity measure to use
    'max_depth': [None, 5, 10, 20],       # Maximum depth of the tree
    'min_samples_split': [2, 5, 10]       # Minimum samples per split
}

# Step 4.2: Create a base DecisionTreeClassifier and GridSearchCV
dt_clf = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(estimator=dt_clf,
                           param_grid=param_grid,
                           scoring='accuracy',
                           cv=5,            # 5-fold cross-validation
                           n_jobs=-1,       # Use all CPU cores
                           verbose=1)

# Step 4.3: Fit the classifier on the training data
grid_search.fit(X_train_scaled, y_train)

# Print the best hyperparameters
print("\nBest Hyperparameters found by GridSearchCV:")
print(grid_search.best_params_)

# The best model after hyperparameter tuning
best_model = grid_search.best_estimator_

###############################################################################
# 5) Model Evaluation
###############################################################################
# Predict on training and test sets
y_pred_train = best_model.predict(X_train_scaled)
y_pred_test = best_model.predict(X_test_scaled)

# Compute metrics
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print("\nTraining Accuracy: {:.4f}".format(train_accuracy))
print("Test Accuracy:     {:.4f}".format(test_accuracy))

print("\nClassification Report (Test Set):")
print(classification_report(y_test, y_pred_test, target_names=["Not Popular", "Popular"]))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_test)
print("Confusion Matrix (Test Set):")
print(cm)

###############################################################################
# 6) Decision Tree Visualization
###############################################################################
# We can visualize the final best tree using sklearn's plot_tree.

plt.figure(figsize=(16, 10))
plot_tree(
    best_model,
    feature_names=features,
    class_names=["Not Popular", "Popular"],
    filled=True,
    max_depth=3,  # For readability, only show first 3 levels
    fontsize=8
)
plt.title("Decision Tree Visualization (Top 3 Levels)")
plt.show()

###############################################################################
# 7) Feature Importance Visualization
###############################################################################
# Decision Trees provide a built-in feature_importances_ attribute.
importances = best_model.feature_importances_
indices = np.argsort(importances)[::-1]
sorted_features = [features[i] for i in indices]

plt.figure(figsize=(10, 6))
plt.bar(range(len(features)), importances[indices])
plt.xticks(range(len(features)), sorted_features, rotation=45, ha='right')
plt.title("Feature Importances in the Tuned Decision Tree")
plt.tight_layout()
plt.show()
```
